## A1. 답안 작성  
답변: A/B Test란, 두 가지 버전(A, B)의 성과를 클릭률이나 전환률과 같은 평가 기준을 기반으로 각 버전의 성과를 비교하는 실험을 의미한다. 이 과정에서 테스트를 위한 각 버전을 설계할 때, 추론 통계적 특성을 활용해 실험을 설계하고 해당 가설을 검정하게 된다.<br>

A/B 테스트 시에 유의할 점들이 여러 가지 있는데, 첫 번째는 **초두효과(Primacy Effect)** 이다. 초두효과는 사용자가 기존 서비스 방식과 다른 서비스를 접했을 때 이에 적응하기 위한 시간이 필요함을 의미한다. A/B 테스트를 위해 처음 수정한 UI에 대하여 사용자가 적응하지 못했다면, 사용자의 혼란으로 인해 초기에는 일시적으로 이용률이 하락하는 등의 부정적인 반응이 있을 수 있다.<br>
초두효과와 반대되는 개념으로 **신기효과(Novelty Effect)** 가 있다. 이는 초두효과와 달리 새로 바뀐 방식에 대하여 사용자들이 호기심을 가져 일시적으로 높은 관심을 보이는 것을 의미한다. A/B test에서 처음 적용된 UI에 사용자들이 관심을 갖고 긍정적인 성과가 측정될 수 있지만, 추후에는 이전과 같거나 더 낮은 수준으로 수렴하는 경우가 있을 수 있다.<br>
위 초두효과와 신기효과는 A/B 테스트로 증명하고자 하는 가설의 실질적인 영향보다는 단기적인 왜곡이 발생할 수 있으므로 이 두 효과에 유의하여 데이터가 안정화 될 때까지 기다리는 과정을 거친 후에 테스트 결과를 해석하고 분석할 필요가 있다.<br>

또한, 테스트 시에 충분한 통계 집단을 대상으로 한 테스트인지를 확인해야 한다. 테스트한 집단의 수가 충분하지 못하다면, 이는 A/B 테스트 결과에 대한 충분한 근거가 되지 못할 수 있기 때문이다. 만일 테스트에 참여한 사용자가 5명 뿐이라면, 이렇게 추정된 결과는 사용자가 100명인 테스트에 비하여 테스트의 결과를 직접적으로 반영한다고 보기 어렵다. 띠라서 신뢰할 수 있는 p-value를 얻고자 적절한 크기의 표본이 필요하다.<br>
마지막으로 테스트 시의 조건은 동일하게 설계해야 한다. 테스트하고자하는 패턴 이외의 UI에 차이가 있거나, 개선이 이루어진다면 이는 순수하게 A/B 테스트로 증명하고자 하는 성과만을 반영했다고 보기 어렵다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:
p1은 검색 미이용자의 기존 상세 진입 전환률인 25%로 지정했고, 30% 이상의 전환률을 목표로 하고잦 하므로 p2는 30%로 지정하였다.

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.

import math
from scipy.stats import norm

alpha = 0.05
beta = 0.8
p1 = 0.25  # 기존 전환률
p2 = 0.3   # 목표 전환률

z_alpha = norm.ppf(1-alpha/2)
z_beta = norm.ppf(beta)


### 공식 계산 부분 ###
numerator = (z_alpha + z_beta)**2 * (p1 * (1 - p1) + p2 * (1 - p2))
denominator = (p1 - p2)**2

n = numerator/denominator

print(f"n = {n}")


#########
```

## Q3-1. 위 가설에 적합한 검정 이름과 왜 해당 검정을 써야하는지 이유도 작성해주세요. (t test와 카이제곱검정 중)

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 카이제곱검정 이고,  

이유는 위 가설은 범주형 데이터인 CVR을 기준으로 하는 가설이기 때문이다.

<br>

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

실제 데이터가 없어서, 저도 필요한 데이터는 가정했답니다.

그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
### A3-4. 코드작성

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [152, 476]  # 대조군(A)
variant_results = [195, 433]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 로그인 유저의 인기순 추천 제공이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 로그인 유저의 인기순 추천 제공의 효과가 통계적으로 증명되지 않았습니다.")




#############
```
