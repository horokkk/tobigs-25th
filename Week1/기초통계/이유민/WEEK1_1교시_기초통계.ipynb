## A1. 답안 작성  
답변:


A/B TEST란 특정 지표를 개선하기 위해 서로 다른 두 가지 버전(A, B안)을 무작위로 나누어 노출하고, 어느 버전이 더 좋은 성과를 내는지 통계적으로 비교 검증하는 실험 방법이다. 이커머스에서 상품 상세 페이지의 '구매하기' 버튼 색상을 파란색(A안)과 초록색(B안)으로 나누어 노출하고, 일정 기간 동안 각 그룹의 클릭률이나 구매 전환율을 비교하면 어떤 색상이 더 높은 전환을 만드는지 검증할 수 있는 것을 예시로 들 수 있다.
A/B TEST 시 유의할 사항은 크게 3가지로 생각할 수 있다. 
1. 실험 결과가 단순 우연이 아닌 '실제 효과'인지 판단하려면 통계적으로 유의한 수준의 표본 수와 실험 기간을 확보해야 한다.
2. 사용자들을 A 그룹과 B 그룹에 무작위로 배정해야 하며, 특정한 특성을 가진 사용자가 한쪽에만 많이 배정되지 않도록 그룹 간 동질성을 확보해야 한다.
3. 처음 본 정보를 더 중요하게 판단하는 경향인 '초두효과'를 고려하여, 실험 과정에서 A안과 B안의 노출 순서를 랜덤하게 섞거나, 두 안이 동일한 맥락에서 비교될 수 있도록 실험을 설계하여야 한다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:

```python
# A2-2 코드작성. 나온 n은 아래 Minimum Sample Size에 기재.


## p1 (A그룹 현재  전환율) = 0.25 (기존 추천 입력률 25% 가정, Observation를 기반으로)
## p2 (B그룹 기대 전환율) = 0.30 (최소 검출 효과 5%p 차이, Hypothesis를 기반으로)
```

```python
import math

p1 = 0.25
p2 = 0.30
z_alpha = 1.96  # Z_{α/2}
z_beta = 0.84   # Z_β
numerator = (z_alpha + z_beta)**2 * ((1-p1)*p1 + (1-p2)*p2)
denominator = (p1 - p2)**2
n = math.ceil(numerator / denominator)
print(f"n: {n}")
```

이유는 해당 실험에서 주지표가 '전환율'이라는 비율이고, 두 그룹(A/B) 간 비율 차이 검증이 목적이기 때문이다.

```python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
```

## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

```python
import numpy as np
from scipy.stats import ttest_ind

n = 1247  

success_A = int(n * 0.25)
fail_A = n - success_A     

success_B = int(n * 0.30)  
fail_B = n - success_B  

print(f"A그룹: 성공 {success_A}, 실패 {fail_A}, CVR {success_A/n:.3f}")
print(f"B그룹: 성공 {success_B}, 실패 {fail_B}, CVR {success_B/n:.3f}")

pA = success_A / n
pB = success_B / n
seA = np.sqrt(pA*(1-pA)/n)
seB = np.sqrt(pB*(1-pB)/n)

t_stat = (pB - pA) / np.sqrt(seA**2 + seB**2)

df = (seA**2 + seB**2)**2 / (seA**4/(n-1) + seB**4/(n-1))
from scipy.stats import t
p_value = t.cdf(-abs(t_stat), df) * 2 

print(f"\nt 통계량: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print(f"\n 귀무가설 기각 (p<{alpha})")
else:
    print(f"\n 귀무가설 기각 못함")
```

**결과해석**
