## A1. 답안 작성  

### A/B TEST 정의
A/B TEST란 두 가지 이상의 버전(A: 대조군, B: 실험군)을 무작위로 나눈 사용자 집단에 동시에 노출시킨 뒤, 사전에 정의한 핵심 지표(KPI)를 통계적으로 비교하여 어느 버전이 더 우수한 성과를 내는지 데이터 기반으로 판단하는 통제된 실험 방법이다.

### 예시
넷플릭스가 콘텐츠 썸네일 이미지의 효과를 비교하는 경우를 들 수 있다. 동일한 영화에 대해 A그룹에는 기존 포스터형 썸네일을, B그룹에는 주인공의 감정 표현이 잘 드러나는 스틸컷 썸네일을 노출시킨 뒤, 콘텐츠 클릭률(CTR)을 비교하여 어떤 이미지가 더 높은 클릭을 유도하는지 검증한다.

### A/B TEST 시 유의사항

1. **초두효과(Primacy Effect)**: 사용자가 기존 UI에 익숙해져 있어 새로운 변경사항에 대해 초기 거부감이나 반대로 신선함에 의한 일시적 행동 변화가 발생할 수 있다. 따라서 실험 초기 데이터만으로 성급하게 판단하지 말고, 충분한 기간 동안 데이터를 수집하여 초두효과가 안정화된 이후의 데이터를 기반으로 분석해야 한다.

2. **무작위 배정(Randomization)**: 대조군과 실험군에 사용자를 무작위로 배정하여 선택 편향(Selection Bias)을 제거해야 한다. 특정 시간대, 지역, 디바이스 등에 치우친 배정은 결과를 왜곡할 수 있다.

3. **충분한 표본 크기**: 통계적으로 유의미한 결과를 도출하기 위해 사전에 필요한 최소 표본 크기를 계산(Power Analysis)하고, 이를 충족해야 한다.

4. **단일 변수 통제**: 한 번에 하나의 변수만 변경하여 테스트해야 결과의 인과관계를 명확히 파악할 수 있다. 여러 변수를 동시에 바꾸면 어떤 변경이 효과를 만들었는지 구분이 불가능하다.

5. **노벨티 효과(Novelty Effect)**: 새로운 UI에 대한 호기심으로 초기에 지표가 일시적으로 상승할 수 있으므로, 이 효과가 사라진 후의 안정적인 데이터를 확보하는 것이 중요하다.

6. **동시 실험(Simultaneous Exposure)**: A/B 두 그룹을 동일한 기간에 동시 노출해야 하며, 시간차를 두면 외부 요인(시즌, 이벤트 등)에 의한 교란이 발생할 수 있다.

## Q2-2. 실험에 필요한 최소 test users 수를 구해봅시다.
유의수준 0.05, 검정력 80% 하에 수행해주세요.  


아래 공식 참고해서, 최소 test users 수를 구하는 코드 작성해주세요.

추가적으로 n은 산출하는데에 가정한 p1, p2도 같이 작성해주세요.

$$n = \frac{(Z_{\alpha/2} + Z_{\beta})^2 \cdot [p_1(1-p_1) + p_2(1-p_2)]}{(p_1 - p_2)^2}$$

## A2-2. 아래 질문에 대한 답과 코드를 채워주세요.

p1, p2는 각각 무엇을 의미하는 확률이고, 어떤 값으로 가정하셨나요?   
또한 위 공식을 보고 n을 구하기 위한 코드를 작성해주세요. 단, 라이브러리는 z값을 구하는 용도의 것만 사용해주세요.


### 답변:
- **p1**: 대조군(A, Control)의 상세 페이지 진입률(CVR). 현재 추천 카드의 CVR이 12%이므로 **p1 = 0.12**로 가정.
- **p2**: 실험군(B, Variant)의 상세 페이지 진입률(CVR). CTA 버튼 추가로 CVR이 15%로 개선될 것으로 기대하여 **p2 = 0.15**로 가정.
- 즉, 3%p의 절대적 향상(상대적으로 약 25% 개선)을 감지하기 위한 최소 표본 크기를 산출한다.

````python
import math
from scipy.stats import norm

# 가정한 전환율
p1 = 0.12  # 대조군 CVR
p2 = 0.15  # 실험군 CVR (기대)

# 유의수준, 검정력
alpha = 0.05
power = 0.80
beta = 1 - power  # 0.20

# Z값 계산 (scipy.stats.norm 사용)
z_alpha_2 = norm.ppf(1 - alpha / 2)  # 양측검정 Z_{alpha/2}
z_beta = norm.ppf(1 - beta)           # Z_{beta}

# 최소 표본 크기 공식
numerator = (z_alpha_2 + z_beta) ** 2 * (p1 * (1 - p1) + p2 * (1 - p2))
denominator = (p1 - p2) ** 2

n = math.ceil(numerator / denominator)

print(f"Z_alpha/2 = {z_alpha_2:.4f}")
print(f"Z_beta    = {z_beta:.4f}")
print(f"p1 = {p1}, p2 = {p2}")
print(f"그룹당 최소 표본 크기 n = {n}명")
print(f"총 최소 표본 크기 = {n * 2}명")
````

## Q3-1. 위 가설에 적합한 검정 이름과 왜 해당 검정을 써야하는지 이유도 작성해주세요. (t test와 카이제곱검정 중)

## A3-1.  검정 이름과 이유를 설명해주세요.

답변: 위 가설에 필요한 검정은 (**카이제곱 검정 (Chi-squared Test)**) 이고,  

이유는 ```종속변수가 '상세 페이지 진입 여부(진입/미진입)'라는 범주형(이진) 변수이며, 독립변수도 '그룹(A/B)'이라는 범주형 변수이기 때문이다. 결국 두 범주형 변수 간의 독립성(연관성)을 검정하는 문제이므로 카이제곱 독립성 검정이 적합하다. t-test는 연속형 종속변수(평균)의 차이를 검정할 때 사용하는 것으로, 전환 여부(0/1)와 같은 비율 데이터를 비교하는 본 실험에는 적합하지 않기 때문``` 이다.

<br>

````python
'''''''''''''''''''''''''''''''''''''''
제가 세운 A/B TEST 설계 실험 코드입니다.
'''''''''''''''''''''''''''''''''''''''

import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 입력 (실제 실험 결과 수치로 변경하세요)
# [상세진입 성공 수, 상세진입 실패 수]
control_results = [11500, 34500]  # 대조군(A)
variant_results = [12190, 33810]  # 실험군(B)

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"P-value: {p_value:.10f}")

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value가 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> 검색바 최상단 노출이 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value가 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> 검색바 위치 변경의 효과가 통계적으로 증명되지 않았습니다.")
````

실제 데이터가 없어서, 저도 필요한 데이터는 가정했답니다.

## Q3-4. 위에서 답한 것을 토대로, 검정을 수행해주세요. (t test와 카이제곱 검정 중)


## A3-4 아래 셀에 코드 작성해주세요. 그리고!!!
그리고 나온 값을 통해 수행하신 a/b test 결과를 설정하신 가설에 맞게 해석해주세요.

````python
import pandas as pd
from scipy.stats import chi2_contingency

# 1. 관측 데이터 (가정)
# 그룹당 약 2,033명씩 배정
# 대조군(A): CVR 12% → 진입 244명, 미진입 1789명
# 실험군(B): CVR 15.8% → 진입 321명, 미진입 1712명
control_results = [244, 1789]   # 대조군(A): [진입 성공, 진입 실패]
variant_results = [321, 1712]   # 실험군(B): [진입 성공, 진입 실패]

# 2. 분할표(Contingency Table) 생성
data = [control_results, variant_results]

# CVR 확인
cvr_a = control_results[0] / sum(control_results)
cvr_b = variant_results[0] / sum(variant_results)
print(f"대조군(A) CVR: {cvr_a:.4f} ({cvr_a*100:.2f}%)")
print(f"실험군(B) CVR: {cvr_b:.4f} ({cvr_b*100:.2f}%)")
print(f"CVR 차이: {(cvr_b - cvr_a)*100:.2f}%p")
print("-" * 50)

# 3. 카이제곱 검정 수행
chi2, p_value, dof, expected = chi2_contingency(data)

# 4. 결과 출력
print(f"카이제곱 통계량: {chi2:.4f}")
print(f"자유도(dof): {dof}")
print(f"P-value: {p_value:.6f}")
print("-" * 50)

alpha = 0.05
if p_value < alpha:
    print(f"결과: P-value({p_value:.6f})가 유의수준 {alpha}보다 작으므로 귀무가설을 기각합니다.")
    print("=> CTA 버튼 추가가 상세 페이지 진입률에 유의미한 변화를 주었습니다.")
else:
    print(f"결과: P-value({p_value:.6f})가 유의수준 {alpha}보다 크므로 귀무가설을 기각할 수 없습니다.")
    print("=> CTA 버튼 추가의 효과가 통계적으로 증명되지 않았습니다.")
````

**결과해석**

답변:

카이제곱 통계량은 11.87, p-value는 약 0.000570으로 유의수준 α=0.05보다 훨씬 작으므로 **귀무가설(H₀: 두 그룹의 CVR이 동일하다)을 기각**한다.

- 대조군(A, 기존 UI)의 상세 페이지 진입률(CVR)은 **12.00%**
- 실험군(B, CTA 버튼 추가)의 상세 페이지 진입률(CVR)은 **15.79%**
- CVR 차이: **+3.79%p** (상대적으로 약 31.6% 개선)

따라서 대립가설(H₁)을 채택하며, **추천 공모전 카드에 "자세히 보기" CTA 버튼을 추가하는 것이 상세 페이지 진입률을 통계적으로 유의미하게 향상**시켰다고 결론 내릴 수 있다. 이는 명확한 행동 유도 요소가 유저의 클릭 행동을 촉진한다는 가설을 지지하며, CTA 버튼 도입을 전체 유저에게 배포하는 것이 합리적인 의사결정이다.
